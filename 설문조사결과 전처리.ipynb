```python
import pandas as pd
```

```python
df = pd.read_csv('전처리.csv')
```

```python
lst = ['날짜','닉네임','성별','나이대','키','몸무게',
       '열_질병_유무','직업','성장환경','여름_겨울',
        '당신상태_1','당신상태_2','당신상태_3','당신상태_4',
      '점수','퀴즈결과','세션_시간(초)']

df.columns = lst
```

```python
df = df.drop(['날짜','닉네임','세션_시간(초)'],axis = 1)
```

```python
df
```

```python
# 해협지 오더 
df = df.iloc[:241,:]
```

```python
df.columns
```

```python
for i in df.columns:
    print(df[i].value_counts())
    print("*"*10)
```

```python
# 나이대 전처리 
df["나이대"] = df["나이대"].str.split('0',expand = True)[0].astype(int)
```

```python
# 키 전처리 
df['키'].replace('150cm 이하', 0 , inplace=True)
df['키'].replace('150 ~ 155cm', 1 , inplace=True)
df['키'].replace('155~160cm', 2 , inplace=True)
df['키'].replace('160~165cm', 3 , inplace=True)
df['키'].replace('165~170cm', 4 , inplace=True)
df['키'].replace('170~175cm', 5 , inplace=True)
df['키'].replace('175cm~180cm', 6 , inplace=True)
df['키'].replace('180cm 이상', 7 , inplace=True)
```

```python
# 열_질병유무 전처리 
df['열_질병_유무'].replace('X', 0 , inplace=True)
df['열_질병_유무'].replace('O', 1 , inplace=True)
```

```python
# 직업 전처리 - 원핫인코딩 
df = pd.get_dummies(df, columns=['직업']);df.head()
```

```python
# 성장환경 - 예상 불쾌지수가 높을수록 높은 라벨 
df['성장환경'].replace('한랭함', 1 , inplace=True)
df['성장환경'].replace('온난함', 2 , inplace=True)
df['성장환경'].replace('고온건조', 3 , inplace=True)
df['성장환경'].replace('고온다습', 4 , inplace=True)
```

```python
# 여름_겨울 - 예상 불쾌지수가 높을수록 높은 라벨 
df['여름_겨울'].replace(df['여름_겨울'][2], 1000 , inplace=True)
df['여름_겨울'].replace(df['여름_겨울'][1], 100 , inplace=True)
df['여름_겨울'].replace('여름 조금 덥긴한데, 그래도 겨울보단 여름!', 10 , inplace=True)
df['여름_겨울'].replace(df['여름_겨울'][4], 1 , inplace=True)
```

```python
# 당신상태_1
df['당신상태_1'].replace(df['당신상태_1'][0], 1 , inplace=True) #뽀송뽀송 이상무 ~ 
df['당신상태_1'].replace(df['당신상태_1'][1], 10 , inplace=True) # 얼굴땀이 송글송글 
df['당신상태_1'].replace(df['당신상태_1'][199], 100 , inplace=True) #등이 흠뻑 젖음
```

```python
# 당신상태_2

df['당신상태_2'].replace(df['당신상태_2'][4], 1 , inplace=True) # 조금 추웠어 !! 벌써 가을이 왔나봐 ~ 
df['당신상태_2'].replace(df['당신상태_2'][3], 300 , inplace=True) # 날씨가 벌써 선선해졌던데? 시원했어 
df['당신상태_2'].replace(df['당신상태_2'][202],500  , inplace=True) # 따뜻하니 좋았어! 적당했어.
df['당신상태_2'].replace(df['당신상태_2'][1], 700 , inplace=True) #좀 덥긴 했는데, 그래도 괜찮았어.  
df['당신상태_2'].replace(df['당신상태_2'][199], 1000 , inplace=True) #너무 더워서 밖에 나가지 않았어\n 에어컨은 무조건 18도 ! 
```

```python
# 당신상태_3
df['당신상태_3'].replace(df['당신상태_3'][0], 1 , inplace=True) # 에어컨 안틀어도 될 날씨같은데, \n가디건을 슥 입어준다 
df['당신상태_3'].replace(df['당신상태_3'][6], 300 , inplace=True) # 선선하니 아주 좋아! 
df['당신상태_3'].replace(df['당신상태_3'][3],500  , inplace=True) # 온도를 조금 내리고 싶은데, \n그래도 불편하진 않아 ! 
df['당신상태_3'].replace(df['당신상태_3'][1], 700 , inplace=True) #조금 덥다고 느껴져서 다른 장소로 이동 !   
```

```python
# 당신상태_4
df['당신상태_4'].replace(df['당신상태_4'][2], 1 , inplace=True) # 반팔 입고 다니고 있어  
df['당신상태_4'].replace(df['당신상태_4'][0], 25 , inplace=True) # 얇은 긴팔 입고 있어
df['당신상태_4'].replace(df['당신상태_4'][3], 75 , inplace=True) # 후드집업 챙겨다니긴 해 
df['당신상태_4'].replace(df['당신상태_4'][200], 100 , inplace=True) # 맨투맨, 후드티 입고 있어

# 이상치 처리 
df['당신상태_4'].replace(df['당신상태_4'][202], 25 , inplace=True) # 얇은 긴팔 티셔츠 한장 -> 얇은 긴팔 입고있어 
df['당신상태_4'].replace(df['당신상태_4'][203], 75 , inplace=True) # 반팔에 아우터도 걸쳐주었지 ! -> 후드집업 챙겨다니긴 해 
```

```python
# 추가 인코딩 
#성별
df['성별'].replace("여자", 0 , inplace=True)
df['성별'].replace("남자", 1 , inplace=True)

#몸무게 
df['몸무게'].replace("40Kg 이하 ", 1 , inplace=True)
df['몸무게'].replace("40 ~ 45Kg ", 10 , inplace=True)
df['몸무게'].replace("45 ~ 50 Kg ", 20 , inplace=True)
df['몸무게'].replace("50 ~ 55 Kg ", 30 , inplace=True)
df['몸무게'].replace("55 ~ 60 Kg ", 40 , inplace=True)
df['몸무게'].replace("60 ~ 65 Kg ", 50 , inplace=True)
df['몸무게'].replace("65 ~ 70 Kg ", 60 , inplace=True)
df['몸무게'].replace("70 ~ 75 Kg ", 70 , inplace=True)
df['몸무게'].replace("75 ~ 80 Kg ", 80 , inplace=True)
df['몸무게'].replace("80 ~ 90 Kg ", 90 , inplace=True)
df['몸무게'].replace("90 ~ 100 Kg ", 100 , inplace=True)
```
